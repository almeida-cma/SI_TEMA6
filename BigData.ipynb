```python
# Importar bibliotecas necessárias
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import io

# Fazer upload do arquivo CSV
uploaded = files.upload()

# Carregar o arquivo CSV
data = pd.read_csv(io.BytesIO(uploaded['random_data.csv']))
print(data.head())

# Aplicar K-means para identificar 5 grupos
kmeans = KMeans(n_clusters=5, random_state=42)
data['Cluster'] = kmeans.fit_predict(data[['Idade', 'Renda_Mensal', 'Quantidade_compras_mensais', 'NPS']])

# Visualizar os centros dos clusters
print(kmeans.cluster_centers_)

# Salvar o DataFrame atualizado em um novo arquivo CSV
data.to_csv('random_data_with_clusters.csv', index=False)

# Resumir as características de cada cluster
cluster_summary = data.groupby('Cluster').mean()
print(cluster_summary)

# Plotar os clusters
sns.pairplot(data, hue='Cluster', palette='viridis')
plt.show()

# Baixar o novo arquivo CSV
files.download('random_data_with_clusters.csv')
```

```python
# Importar bibliotecas necessárias
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import io

# Fazer upload do arquivo CSV
uploaded = files.upload()

# Carregar o arquivo CSV
data = pd.read_csv(io.BytesIO(uploaded['random_data.csv']))
print(data.head())

# Aplicar K-means para identificar 5 grupos
kmeans = KMeans(n_clusters=5, random_state=42)
data['Cluster'] = kmeans.fit_predict(data[['Idade', 'Total_compras_mensal', 'Qtd_compras_mensal', 'Renda_mensal']])

# Visualizar os centros dos clusters
print(kmeans.cluster_centers_)

# Salvar o DataFrame atualizado em um novo arquivo CSV
data.to_csv('random_data_with_clusters.csv', index=False)

# Resumir as características de cada cluster
cluster_summary = data.groupby('Cluster').mean()
print(cluster_summary)

# Plotar os clusters
sns.pairplot(data, hue='Cluster', palette='viridis')
plt.show()

# Baixar o novo arquivo CSV
files.download('random_data_with_clusters.csv')
```
